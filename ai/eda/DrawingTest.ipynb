```python
import ast
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def draw(idx :int, data):
    points = ast.literal_eval(data.iloc[idx].drawing)
    print(idx)
    for x,y in points:
        plt.plot(y,x, marker='.')
```

```python
def delete_datasets(file_path : str):
    import os
    from collections import defaultdict

    with open(file_path, 'r') as f:
        not_in_class = defaultdict(int)
        file_lists = [i for i in os.listdir() if '.csv' in i]

        for ff in file_lists:
            not_in_class[ff] +=1

        for line in f:
            _key = line.rstrip('\n').rstrip(' ')
            not_in_class[f'{_key}.csv'] +=1

        cnt = 0
        for k,v in not_in_class.items():
            if v != 2:
                try:
                    os.remove(k)
                    cnt+=1
                except:
                    pass
        print(f'delete {cnt}')
```

```python
datas = pd.read_csv('wristwatch.csv')
datas = datas[datas.recognized == True]
draw(random.randrange(1, len(datas)), datas)
```
